```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
dbs = pd.read_csv('databases.csv', sep = '\t')
```

```python
dbs.head(3)
```

```python
F = nx.Graph()
F.add_edges_from([(i, j) for i, j in zip(dbs.Category, dbs.Dbs)])
```

```python
POSICIONES = nx.spring_layout(F, seed = 12)
POSICIONES = nx.spring_layout(F, k = 0.1, iterations=100, threshold=0.001, seed= 123)

POS = {}
POS_LAB = {}
aumento_lab_y = 0
for i in F.nodes:
    POS[i] = POSICIONES[i]
    POS_LAB[i] = POSICIONES[i]
```

```python
fig,ax = plt.subplots(figsize=(5, 15))
nx.draw_networkx_nodes(F, POS, node_color= 'salmon',
                           alpha= 1,node_size =  100,
                           zorder = 2)
nx.draw_networkx_edges(F, POS, width = 3,alpha= 0.3,edge_color=  'gray',style='-')
for i in POS_LAB:
    if i in ['Protein Sequences','Protein Domains and Families','Protein Structures',
             'Protein Modifications','Protein Interactions','Protein Localization',
             'Protein tools',
             'Genes','RNA-Specific Sequences','Software/Web Servers','Genomes',
             'Other Nucleotide Sequence Databases','Databases/Resources']:
        ax.annotate(i, xy=POS_LAB[i], xycoords='data',  rotation=270, zorder = 2, 
                    alpha = 0.8, fontweight='bold',
                    xytext=POS_LAB[i], textcoords='data', ha="left", va = 'center', color = 'black',
                    size = 8)
    if i in ['Proteins', 'Nucleic Acids']:
        ax.annotate(i, xy=POS_LAB[i], xycoords='data',  rotation=270, zorder = 2, 
                    alpha = 0.5, fontweight='bold',
                    xytext=POS_LAB[i], textcoords='data', ha="left", va = 'center', color = 'black',
                    size = 30)
ax.axis('off')
plt.savefig('interactionsDBs.png',dpi = 900, bbox_inches='tight', transparent=True)
plt.show()
```
