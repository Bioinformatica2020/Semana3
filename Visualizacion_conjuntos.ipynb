## 1. Construcción de conjuntos

```python
set1 = set(list(range(1, 101)))
```

```python
set2 = set(list(range(60, 200)))
```

```python
set3 = set(list(range(80, 180)))
```

## 2. Cargamos los módulos para diagramas de Venn

## Diagramas de Venn 2 conjuntos

```python
# llamamos funciones del módulo matplotlib-venn
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
import matplotlib.pyplot as plt
```

```python
# (Ab, aB, AB)
grupos2 = ('10', '01', '11')
grupos2
```

```python
plt.figure(figsize=(5, 5))

v = venn2_unweighted(subsets=[set1, set2], set_labels=('Set1', 'Set2'))
c = venn2_circles(subsets = (1, 1, 1),linestyle='--', linewidth= 2, color='black')

for i in grupos2:
    v.get_patch_by_id(i).set_alpha(0)
    v.get_label_by_id(i).set_text(i)
    v.get_label_by_id(i).set_fontsize(15)

v.get_label_by_id('A').set_fontsize(25)
v.get_label_by_id('B').set_fontsize(25)

plt.show()
```

```python
plt.figure(figsize=(5, 5))

v = venn2_unweighted(subsets=[set1, set2], set_labels=('Set1', 'Set2'))
c = venn2_circles(subsets = (1, 1, 1),linestyle='--', linewidth= 2, color='black')


v.get_patch_by_id('11').set_alpha(0)
v.get_label_by_id('11').set_fontsize(25)
v.get_label_by_id('11').set_fontweight('bold')


v.get_label_by_id('10').set_fontsize(20)
v.get_label_by_id('01').set_fontsize(20)
v.get_label_by_id('A').set_fontsize(25)
v.get_label_by_id('B').set_fontsize(25)

# si quieres salvar el gráfico activa la siguiente línea, y vuelve a correr la celda
#plt.savefig('../salidas/Conjuntos2.png',dpi = 600, bbox_inches='tight')

plt.show()
```

## Diagramas de Venn 3 conjuntos

```python
from matplotlib_venn import venn3, venn3_circles, venn3_unweighted
```

```python
# (Abc, aBc, ABc, abC, AbC, aBC, ABC)
grupos3 = ('100', '010', '110', '001', '101', '011', '111') 
```

```python
plt.figure(figsize=(5, 5))

v = venn3_unweighted(subsets=[set1, set2, set3], set_labels=('Set1', 'Set2', 'Set3'))
c = venn3_circles(subsets = (1, 1, 1, 1, 1, 1, 1), linestyle='--', linewidth= 2, color='black')

for i in grupos3:
    v.get_patch_by_id(i).set_alpha(0)
    v.get_label_by_id(i).set_text(i)
    v.get_label_by_id(i).set_fontsize(15)

v.get_label_by_id('A').set_fontsize(15)
v.get_label_by_id('B').set_fontsize(15)
v.get_label_by_id('C').set_fontsize(15)

plt.show()
```

```python
plt.figure(figsize=(5, 5))

v = venn3_unweighted(subsets=[set1, set2, set3], set_labels=('Set1', 'Set2', 'Set3'))
c = venn3_circles(subsets = (1, 1, 1, 1, 1, 1, 1), linestyle='--', linewidth= 2, color='black')


v.get_patch_by_id('111').set_alpha(0)
v.get_label_by_id('111').set_fontsize(25)
v.get_label_by_id('111').set_fontweight('bold')

for i in grupos3:
    v.get_label_by_id(i).set_fontsize(20)
    

v.get_label_by_id('A').set_fontsize(25)
v.get_label_by_id('B').set_fontsize(25)
v.get_label_by_id('C').set_fontsize(25)

# si quieres salvar el gráfico activa la siguiente línea, y vuelve a correr la celda
#plt.savefig('../salidas/Conjuntos3.png',dpi = 600, bbox_inches='tight')

plt.show()
```
